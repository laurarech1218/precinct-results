```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
df = pd.read_csv('downloads\\precinct_level_election_results.csv')
df.head()
df.shape
```

```python
df['ballot_measure_support_proportion'] = (df['votes_for_ballot_measure'])/(df['total_ballot_measure_votes'])
df['ballot_measure_support_proportion']
```

```python
df.loc[df['ballot_measure_support_proportion']>1]
```

```python
df
```

```python
df['for_candidate_I'] = (df['votes_for_candidate_I'])/(df['total_ballots'])
df['for_candidate_I']
```

```python
df.loc[df['for_candidate_I']>1]
```

```python
df['for_candidate_U'] = (df['votes_for_candidate_U'])/(df['total_ballots'])
df['for_candidate_U']
```

```python
df.loc[df['for_candidate_U']>1]
```

```python
df['for_president']=(df['votes_for_president'])/(df['total_ballots'])
df['for_president']
```

```python
df.loc[df['for_president']>1]
```

```python
df2=df[df['for_president']<1]
df2.shape
```

```python
import statsmodels.api as sm
```

```python
X = df2[['millenial_proportion','for_candidate_I','for_candidate_U','partisan_score','turnout_score']] 
Y = df2['ballot_measure_support_score']

# with statsmodels
X = sm.add_constant(X) # adding a constant
 
model = sm.OLS(Y, X).fit()
predictions = model.predict(X) 
 
print_model = model.summary()
print(print_model)
```
